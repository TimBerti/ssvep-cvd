```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('../images/style_sheet.mplstyle')
```

```python
df = pd.read_csv('../data/linss2_10e_fine.csv', header=None, names=['wavelength', 'R', 'G', 'B'])
df.fillna(0, inplace=True)

df = df[df['wavelength'] < 750]

fig, ax = plt.subplots()
ax.plot(df['wavelength'], df['B'], label='S', color='blue')
ax.plot(df['wavelength'], df['G'], label='M', color='green')
ax.plot(df['wavelength'], df['R'], label='L', color='red')

ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Normalized Response')
ax.legend()
plt.savefig('../images/cone_activation.pdf', format='pdf')
plt.show()
```
