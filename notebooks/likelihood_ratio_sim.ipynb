```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
a_0 = 0.2797528238162667
b_0 = 1.1641546902903654
a_1 = 1.9997027282059479
b_1 = 0.4650707013190572

def likelihood_ratio(sample):
    h_0 = np.sum(beta.logpdf(sample, a_0, b_0))
    h_1 = np.sum(beta.logpdf(sample, a_1, b_1))
    return np.exp(h_1 - h_0)

samples = beta.rvs(a_0, b_0, size=15 * 1000000).reshape(-1, 15)
ratios = np.apply_along_axis(likelihood_ratio, 1, samples)
```

```python
plt.hist(ratios, bins=100, density=True, range=(0, np.percentile(ratios, 50)));
```
