```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import iqr
```

```python
file_path = '../data/2024-01-12/Geri_Deuteranomaly_0.0_multicolor-deuteranomaly-series_frequency_16.5_tile-size144_8e3e867f-c88b-4fac-afa4-a77c93890989.csv'
df = pd.read_csv(file_path, skiprows=10, header=None)
timestamp = df.to_numpy()[:, 0].astype(float)
```

```python
time_diff = np.diff(timestamp)
range = iqr(time_diff, rng=(2, 98))
median = np.median(time_diff)

print(timestamp[-1] - timestamp[0])
print(median - range, median + range)

plt.plot(time_diff)
plt.xlabel('Sample')
plt.ylabel('Timestep (s)');
```

```python
plt.plot(np.cumsum(time_diff - median))
plt.xlabel('Sample')
plt.ylabel('Cumulative Timestep Deviation (s)');
```
